## **4.5. Machine Learning Models**

- **Logistic Regression (LR)**:
  Logistic regression is a classification algorithm that models the probability of a binary outcome based on one or more predictor variables using a logistic function.
  
- **Support Vector Machines (SVM)**:
  SVM finds a hyperplane that best separates the data into classes by maximizing the margin between the closest points of the classes (support vectors).

- **Stochastic Gradient Descent (SGD)**:
  An iterative optimization algorithm used for minimizing loss functions in linear classifiers and regressors, particularly useful for large datasets.

- **Decision Trees**:
  A tree-based algorithm where each internal node represents a decision on a feature, each leaf node represents an outcome, and paths from root to leaf represent classification rules.

- **K-Nearest Neighbors (KNN)**:
  A lazy learning algorithm that classifies a data point based on the majority class of its k-nearest neighbors in the feature space.
  
- **Naive Bayes (NB)**:
  A probabilistic classifier based on Bayes' theorem, assuming strong independence between features.

- **Bagging (Bootstrap Aggregating)**:
  Combines predictions from multiple models (e.g., Random Forest) to reduce variance and improve accuracy.

- **Boosting**:
  Combines weak learners sequentially to form a strong learner by minimizing errors iteratively (e.g., Extreme Gradient Boosting - XGBoost).

## **4.6. Model Performance Metrics**

- **Confusion Matrix**:
  - Sensitivity (Recall): True Positive Rate
  - Specificity: True Negative Rate
  - Accuracy: Overall correctness
  - Precision: Positive Predictive Value
  - F1-Score: Harmonic mean of Precision and Recall
  - AUC: Area Under the ROC Curve

- **K-Fold Cross-Validation**: Splits data into k subsets; trains on k-1 and tests on the remaining one iteratively to ensure robust performance evaluation.

- **Model Run Statistics**:
  - Time Taken
  - Memory Used
  - Inherent Complexity

````python
# Handling Missing Data
def missing_data_summary(df):
    """
    Creates a DataFrame summarizing missing data statistics.

    Args:
        df: The input DataFrame.

    Returns:
        A DataFrame with missing data count and percentage for each variable and row.
    """
    # Column-wise summary
    missing_count_col = df.isnull().sum()
    missing_percentage_col = (missing_count_col / len(df)) * 100
    missing_summary_col = pd.DataFrame({
        'Missing Count': missing_count_col,
        'Missing Percentage': missing_percentage_col
    })
    # Row-wise summary
    missing_count_row = df.isnull().sum(axis=1)
    missing_percentage_row = (missing_count_row / len(df.columns)) * 100
    missing_summary_row = pd.DataFrame({
        'Missing Count': missing_count_row,
        'Missing Percentage': missing_percentage_row
    }, index=df.index)  # Use original DataFrame index

    return missing_summary_col, missing_summary_row

# Calculate missing data summary for categorical subset
sy45_missing_summary_cat_col, sy45_missing_summary_cat_row = missing_data_summary(sy45_ds_cat)

# Calculate missing data summary for non-categorical subset
sy45_missing_summary_noncat_col, sy45_missing_summary_noncat_row = missing_data_summary(sy45_ds_noncat)

# Drop columns and rows with more than 50% missing data for categorical subset
sy45_ds_cat_treated = sy45_ds_cat.dropna(thresh=len(sy45_ds_cat) * 0.5, axis=1)
sy45_ds_cat_treated = sy45_ds_cat_treated.dropna(thresh=len(sy45_ds_cat_treated.columns) * 0.5, axis=0)

# Drop columns and rows with more than 50% missing data for non-categorical subset
sy45_ds_noncat_treated = sy45_ds_noncat.dropna(thresh=len(sy45_ds_noncat) * 0.5, axis=1)
sy45_ds_noncat_treated = sy45_ds_noncat_treated.dropna(thresh=len(sy45_ds_noncat_treated.columns) * 0.5, axis=0)

# Calculate total variables and rows dropped
sy45_dropped_variables_cat = len(sy45_ds_cat.columns) - len(sy45_ds_cat_treated.columns)
sy45_dropped_rows_cat = len(sy45_ds_cat) - len(sy45_ds_cat_treated)
sy45_dropped_variables_noncat = len(sy45_ds_noncat.columns) - len(sy45_ds_noncat_treated.columns)
sy45_dropped_rows_noncat = len(sy45_ds_noncat) - len(sy45_ds_noncat_treated)

sy45_total_dropped_variables = sy45_dropped_variables_cat + sy45_dropped_variables_noncat
sy45_total_dropped_rows = sy45_dropped_rows_cat + sy45_dropped_rows_noncat

# Print missing data summaries
print("Categorical Missing Data Summary (Columns):")
print(sy45_missing_summary_cat_col)
print("\nCategorical Missing Data Summary (Rows):")
print(sy45_missing_summary_cat_row)
print("\nNon-Categorical Missing Data Summary (Columns):")
print(sy45_missing_summary_noncat_col)
print("\nNon-Categorical Missing Data Summary (Rows):")
print(sy45_missing_summary_noncat_row)

# Print total dropped variables and rows
print(f"\nTotal variables dropped: {sy45_total_dropped_variables}")
print(f"Total rows dropped: {sy45_total_dropped_rows}")
````

````python
# Calculate decision thresholds for Quantity and Weight between each pair of classes
def calculate_threshold(mu1, mu2, var1, var2):
    if var1 != var2:
        return (mu1 * var2 - mu2 * var1) / (var2 - var1)
    else:
        return None  # No threshold if variances are equal

# Extract means and variances
means = stats_df.loc[:, stats_df.columns.str.contains("_mean")]
variances = stats_df.loc[:, stats_df.columns.str.contains("_variance")]

# Calculate thresholds between each pair of classes for each feature
thresholds = {}
for feature in means.index:
    thresholds[feature] = {}
    for (cls1, cls2) in [(0, 1), (0, 2), (1, 2)]:
        mu1, mu2 = means[f"{cls1}.0_mean"][feature], means[f"{cls2}.0_mean"][feature]
        var1, var2 = variances[f"{cls1}.0_variance"][feature], variances[f"{cls2}.0_variance"][feature]
        thresholds[feature][f"Class_{cls1}_vs_Class_{cls2}"] = calculate_threshold(mu1, mu2, var1, var2)

# Convert thresholds to a DataFrame for readability
thresholds_df = pd.DataFrame(thresholds)
thresholds_df
````

## **Model Run Statistics**

### **Time Taken**

````python
import time
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def time_model_run(model, X_train, y_train):
    start_time = time.time()
    model.fit(X_train, y_train)
    end_time = time.time()
    return end_time - start_time

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55045)
log_reg_time = time_model_run(log_reg_model, sy45_X_train, sy45_y_train)
print(f"Logistic Regression runtime: {log_reg_time:.4f} seconds")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55045)
svm_time = time_model_run(svm_model, sy45_X_train, sy45_y_train)
print(f"SVM runtime: {svm_time:.4f} seconds")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55045)
sgd_time = time_model_run(sgd_model, sy45_X_train, sy45_y_train)
print(f"SGD runtime: {sgd_time:.4f} seconds")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55045)
dt_time = time_model_run(dt_model, sy45_X_train, sy45_y_train)
print(f"Decision Tree runtime: {dt_time:.4f} seconds")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_time = time_model_run(knn_model, sy45_X_train, sy45_y_train)
print(f"KNN runtime: {knn_time:.4f} seconds")

# Naive Bayes
nb_model = GaussianNB()
nb_time = time_model_run(nb_model, sy45_X_train, sy45_y_train)
print(f"Naive Bayes runtime: {nb_time:.4f} seconds")

# Random Forest
rf_model = RandomForestClassifier(random_state=55045)
rf_time = time_model_run(rf_model, sy45_X_train, sy45_y_train)
print(f"Random Forest runtime: {rf_time:.4f} seconds")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55045)
xgb_time = time_model_run(xgb_model, sy45_X_train, sy45_y_train)
print(f"XGBoost runtime: {xgb_time:.4f} seconds")
````

### **Observations**

#### **Fastest:**
- **KNN** and **Naive Bayes** are the fastest models, taking less than **0.01 seconds** for execution.

#### **Slowest:**
- **Random Forest** is the slowest model, taking approximately **1.255 seconds**, which may be a limitation for real-time applications.

#### **Moderate Runtimes:**
- **Logistic Regression**, **SGD**, and **Decision Tree** have moderate runtimes ranging between **0.02 to 0.05 seconds**.
- These models balance speed and computational complexity effectively.

#### **Relatively Slow:**
- **SVM** takes approximately **0.67 seconds**, and **XGBoost** takes **0.31 seconds**, making them relatively slower compared to the other models.

---

### **Insights**

1. **KNN and Naive Bayes:**
   - **Use Case:** Best suited for quick analysis or scenarios where runtime is a critical factor.
   - **Limitation:** While fast, these models may trade off accuracy compared to more complex models.

2. **Logistic Regression, SGD, and Decision Tree:**
   - **Use Case:** Offer a good balance between speed and predictive power, making them ideal for moderate-sized datasets.
   - **Strength:** These models are efficient for both training and inference tasks.

3. **Random Forest:**

###**Memory Utilized**

````python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(model, X, y):
    tracemalloc.start()
    model.fit(X, y)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55045)
log_reg_memory = get_model_memory_usage(log_reg_model, sy45_X_train, sy45_y_train)
print(f"Logistic Regression memory usage: {log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55045)
svm_memory = get_model_memory_usage(svm_model, sy45_X_train, sy45_y_train)
print(f"SVM memory usage: {svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55045)
sgd_memory = get_model_memory_usage(sgd_model, sy45_X_train, sy45_y_train)
print(f"SGD memory usage: {sgd_memory:.2f} MB")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55045)
dt_memory = get_model_memory_usage(dt_model, sy45_X_train, sy45_y_train)
print(f"Decision Tree memory usage: {dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_memory = get_model_memory_usage(knn_model, sy45_X_train, sy45_y_train)
print(f"KNN memory usage: {knn_memory:.2f} MB")

# Naive Bayes
nb_model = GaussianNB()
nb_memory = get_model_memory_usage(nb_model, sy45_X_train, sy45_y_train)
print(f"Naive Bayes memory usage: {nb_memory:.2f} MB")

# Random Forest
rf_model = RandomForestClassifier(random_state=55045)
rf_memory = get_model_memory_usage(rf_model, sy45_X_train, sy45_y_train)
print(f"Random Forest memory usage: {rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55045)
xgb_memory = get_model_memory_usage(xgb_model, sy45_X_train, sy45_y_train)
print(f"XGBoost memory usage: {xgb_memory:.2f} MB")
````

### **Observations**

#### **Highest Memory Usage:**
- **XGBoost** has the highest memory usage, requiring **0.90 MB**, making it the most resource-intensive model.

#### **Lowest Memory Usage:**
- **Naive Bayes** has the lowest memory usage, requiring only **0.19 MB**, making it ideal for memory-sensitive tasks.

#### **Moderate Memory Usage:**
- **Logistic Regression** (**0.38 MB**), **SVM** (**0.29 MB**), **Decision Tree** (**0.27 MB**), **KNN** (**0.25 MB**), and **Random Forest** (**0.50 MB**) have moderate memory usage, suitable for most applications where memory is not a major constraint.

#### **Relatively Low Memory Usage:**
- **SGD** has relatively low memory usage at **0.20 MB**, making it efficient for resource-constrained environments.

---

### **Insights**

1. **XGBoost:**
   - **Use Case:** Best suited for tasks where accuracy is prioritized over memory efficiency.
   - **Limitation:** Requires careful resource management due to its high memory usage.

2. **Random Forest:**
   - **Use Case:** Offers high accuracy with moderate memory usage but less efficient than simpler models like Naive Bayes or SGD.

3. **Logistic Regression, SVM, Decision Tree, and KNN:**
   - **Use Case:** These models strike a balance between predictive power and memory efficiency, making them versatile choices for various tasks.
